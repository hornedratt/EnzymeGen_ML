```python
#порезали нулевые последовательности
zero_seq_1 = [i for i in zero_seq_main[1]]
# zero_seq = zero_seq_main[1]
zero_seq_2 = [i for i in zero_seq_main[3]]
# zero_seq = zero_seq_main[1]
zero_seq_1[75]
```

```python
#выбрасываем плохо пропарсенные строки (где мутация длиннее чем 2 буквы + трехзначное число)
clean_test = test
empty = test.isnull()
for i in range(len(test.index)):
    for j in range(1, index_end_mut + 1):
        if empty.at[i, j] == False:
            if len(test.at[i, j])>5:
                clean_test = clean_test.drop(index=[i])
                break
        else: continue
clean_test.index = [i for i in range(len(clean_test.index))]
clean_test
```

```python
#чекним, какие ошибки в парсинге
wrong_stuff = []

condition = lambda x: isinstance(x, str) and not(x[len(x) - 1].isalpha() and x[0].isalpha())

# Iterate over each element in the DataFrame
for i in range(len(clean_test.index)):
    for j in range(1, index_end_mut + 1):
        if condition(clean_test.at[i, j]):
            print(clean_test.at[i, j])
            wrong_stuff.append(clean_test.at[i, 0])
wrong_stuff
```

```python
condition = lambda val: isinstance(val, str) and (val[len(val) - 1].isdigit() or val[0].isdigit())
#Меняем 1 на I
for i in range(len(clean_test.index)):
    for j in range(1, index_end_mut + 1):
        if condition(clean_test.at[i, j]) and clean_test.at[i, j][0].isdigit():
            clean_test.at[i, j] = 'I'+ clean_test.at[i, j][1:]
        if condition(clean_test.at[i, j]) and clean_test.at[i, j][len(clean_test.at[i, j])-1].isdigit():
            clean_test.at[i, j] = clean_test.at[i, j][:len(clean_test.at[i, j])-1]+'I'

# Apply the function to the DataFrame
clean_test
```

```python
#отбираем мутации для второй последовательности, идем по первой мутации, если аминокимлота в указанном месте
#совпадает с соответствующей аминокислотой в нулевой последовательности - делаем вывод, что мутация относится
# к ней,плюс сразу создаем df для первой последовательности (clean_test_1)
indices_for_2_seq = []
for i in range (len(clean_test.index)):
    index = clean_test.at[i, 1][1]
    for j in range(2, len(clean_test.at[i, 1]) - 1):
        index = index + clean_test.at[i, 1][j]
    index = int(index)
    if clean_test.at[i, 1][0] != zero_seq_1[index - 1]:
        indices_for_2_seq.append(clean_test.at[i, 0])

clean_test = clean_test[~clean_test[0].isin(indices_for_2_seq)]
indices_for_2_seq
```

```python
whole_test = []
```

```python
empty = clean_test.isnull()
for k in clean_test.index:
    mut_seq = zero_seq_1
    for i in range(1, index_end_mut + 1):
        mut = clean_test.at[k, i]
        if empty.at[k, i] == False:
            index = mut[1]
            for j in range(2, len(mut) - 2):
                index = index + mut[j]
            index = int(index)
            mut_seq[index - 1] = mut[len(mut) - 1]
        else: continue
    mut_seq = ''.join(mut_seq)
    whole_test.append(CustomDataSet(sequence=(str(clean_test.at[k, 0]), mut_seq),
                                    cloatting=clean_test.at[k, index_cloatting],
                                    protalitic=clean_test.at[k, index_protalitic],
                                    CP=clean_test.at[k, CP]))
```

```python
import pickle
with open('whole_test.pkl', 'wb') as file:
    pickle.dump(whole_test, file)
```
